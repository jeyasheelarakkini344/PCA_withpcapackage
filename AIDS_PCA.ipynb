```python
from sklearn.preprocessing import StandardScaler
X=d1
sc=StandardScaler()
X_scaled=sc.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=6)
pca.fit(X_scaled)
```

```python
print("Eigenvalues")
print(pca.explained_variance_)
print("Eigenvectors")
print(pca.components_)
```
